**Ejercicio 1.** Programa para validar una contraseña que debe tener máximo 10 caracteres y mínimo 8. La contraseña debe contener al menos 3 letras minúsculas, 1 letra mayúscula y 3 números.

```python
while(True):
  contra=input("Ingrese la contraseña: ")
  if(len(contra)<8 or len(contra)>10):
    print("Contraseña inválida, debe tener entre 8 y 10 carcteres.")
  else:
    min_cont=0
    may_cont=0
    num_cont=0
    for i in range(len(contra)):
      if(contra[i].islower()):
        min_cont+=1
      elif(contra[i].isupper()):
        may_cont+=1
      elif(contra[i].isdecimal()):
        num_cont+=1
    if(min_cont >2 and may_cont > 0 and num_cont>2):
      break
    print("Contraseña inválida, debe tener al menos 1 mayúscula, 3 minúsculas y 3 números.")
print("Contraseña aceptada.")
```

**Ejercicio 2.** Programa para ingresar 10 números flotantes en el rango de (0,1) (ambos incluidos) en una lista. Debemos imprimir el valor menor y mayor ingresados. Además, crearemos otra lista con los elementos que superen el valor promedio de la primera lista e imprimirlo.


```python
print("Debe ingresar 10 números que estén en el intervalo (0,1).")
lista=[]
while(len(lista)<10):
  num=float(input("Ingrese un número, faltan "+str(10-len(lista))+" por ingresar: "))
  if(num<1 and num>0):
    lista.append(num)
  else:
    print("El número debe estar en el intervalo (0,1).")

print("\nLa lista final es: ", lista)
print("El menor valor ingresado fue: ", min(lista))
print("El mayor valor ingresado fue: ", max(lista))

suma=0
for i in lista:
  suma+=i
promedio=suma/len(lista)

print("El promedio de la lista es: ", promedio)

lista_mayores=[]
for i in lista:
  if(i>promedio):
    lista_mayores.append(i)

print("La lista de números que superan el promedio de la lista son: ", lista_mayores)
```
